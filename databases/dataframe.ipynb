```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
```

```python
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="x",
    database="data_science"
)
print(mydb)
```

```python
mycursor = mydb.cursor()
```

## 1.- Graficar el número de SNIs por carrera

```python
sql = "SELECT c.nombre AS Carrera, COUNT(p.carrera) AS numeroSni FROM profesores AS p INNER JOIN (SELECT * FROM sni GROUP BY(correoProfesor)) AS sni ON p.correo=sni.correoProfesor INNER JOIN carreras AS c ON p.carrera=c.codigoCarrera GROUP BY(p.carrera)"
datos = pd.read_sql(sql,mydb)
datos
```

```python
plt.figure(figsize=(16,8))
plt.bar(datos['Carrera'],datos['numeroSni'])
plt.title("UTM")
plt.title("Gráfica del número de SNIs por carrera")
plt.xticks(rotation=45, ha="right", fontsize=10)
```

## 2.- Graficar el número de articulos por carrera

```python
sql = "SELECT c.nombre AS Carrera, COUNT(p.carrera) AS numeroArticulos FROM profesoryarticulo AS pa INNER JOIN articulos AS a ON pa.numero=a.numero INNER JOIN profesores AS p ON p.correo=pa.correo INNER JOIN carreras AS c ON c.codigoCarrera=p.carrera GROUP BY(p.carrera)"
datos = pd.read_sql(sql,mydb)
datos
```

```python
plt.figure(figsize=(16,8))
plt.bar(datos['Carrera'],datos['numeroArticulos'])
plt.title("UTM")
plt.title("Gráfica del número de articulos por carrera")
plt.xticks(rotation=45, ha="right", fontsize=10)
```

## 3.- Graficar el número de actividades por carrera

```python
sql = "SELECT c.nombre AS Carrera,COUNT(c.codigoCarrera) AS numeroActividades FROM actividades AS a INNER JOIN profesores AS p ON a.idProfesor=p.id INNER JOIN carreras AS c ON p.carrera=c.codigoCarrera GROUP BY(c.codigoCarrera)"
datos = pd.read_sql(sql,mydb)
datos
```

```python
plt.figure(figsize=(16,8))
plt.bar(datos['Carrera'],datos['numeroActividades'])
plt.title("UTM")
plt.title("Gráfica del número de actividades por carrera")
plt.xticks(rotation=45, ha="right", fontsize=10)
```

## 4.- Graficar el número de tesistas por carrera

```python
sql = "SELECT C.nombre AS Carrera,COUNT(C.codigoCarrera) AS numeroTesistas FROM profesores AS P INNER JOIN  carreras AS C ON P.carrera=C.codigoCarrera INNER JOIN profesorytesis AS PT ON PT.correo=P.correo INNER JOIN tesistas AS T ON PT.idTesis=T.id GROUP BY C.codigoCarrera"
datos = pd.read_sql(sql,mydb)
datos
```

```python
plt.figure(figsize=(16,8))
plt.bar(datos['Carrera'],datos['numeroTesistas'])
plt.title("UTM")
plt.title("Gráfica del número de tesistas por carrera")
plt.xticks(rotation=45, ha="right", fontsize=10)
```

## 5.- Graficar el número de tesistas por profesor de una carrera determinada

```python
sql = "SELECT P.nombre profesor,C.nombre AS carrera,COUNT(PT.idTesis) AS numeroTesistas FROM profesorytesis AS PT INNER JOIN profesores AS P ON PT.correo=P.correo  INNER JOIN carreras AS C ON P.carrera=C.codigoCarrera GROUP BY(PT.idTesis) ORDER BY(Carrera) LIMIT 20"
datos = pd.read_sql(sql,mydb)
datos
```

```python
pivot_consulta = datos.pivot(index="profesor",columns="carrera",values="numeroTesistas").fillna(0)
```

```python
pivot_consulta
```

```python
pivot_consulta.plot(kind="bar",figsize=(12,8))
plt.title("Número de tesistas por profesor de una carrera determinada",fontsize=16)
plt.xlabel("Profesor",fontsize=12)
plt.ylabel("Numero Tesistas",fontsize=12)
plt.xticks(rotation=45,ha="right",fontsize=10)
plt.legend(title="Carrera",fontsize=10)
plt.grid(axis="y",linestyle="--",alpha=0.7)
plt.show()
```

## 6.- Graficar el número de materias por profesor de una carrera determinada

```python
sql = "SELECT P.nombre AS profesor,C.nombre AS carrera,COUNT(P.nombre) AS numeroMaterias FROM profesorymateria AS PT INNER JOIN profesores AS P ON PT.profesor=P.correo INNER JOIN carreras AS C ON C.codigoCarrera=P.carrera GROUP BY(P.nombre) LIMIT 15"
datos = pd.read_sql(sql,mydb)
datos
```
