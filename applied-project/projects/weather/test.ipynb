```python
import pandas as pd
```

```python
# Crear los dos DataFrames con las mismas columnas
df1 = pd.DataFrame(
    {"Nombre": ["Alice", "Bob"], "Edad": [25, 30], "Ciudad": ["Madrid", "Barcelona"]}
)

df2 = pd.DataFrame(
    {
        "Nombre": ["Alice", "Bob", "Diana"],
        "Edad": [25, 30, 27],
        "Ciudad": ["Madrid", "Barcelona", "Zaachila"],
    }
)

# Combinar los DataFrames
df_combinado = pd.concat([df1, df2], ignore_index=True).drop_duplicates()

# Mostrar el resultado
print(df_combinado)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
# Combinar los DataFrames usando merge para obtener las filas comunes (equivalente a eliminar duplicados)
df_combinado = pd.merge(df1, df2, on=['Nombre', 'Edad', 'Ciudad'])
```

```python
pd.merge(df1,df2,how="right")
```
