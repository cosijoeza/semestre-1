# Lección 17 - Dummies

El objetivo es llevar variables de texto a valores númericos

```python
import pandas as pd
import numpy as np
```

## Variables Discretas

```python
movies = pd.read_csv("../data/series-movies.csv")
```

```python
movies.head()
```

```python
movies = movies.loc[:,["Series_Title","Released_Year","Genre"]]
movies
```

```python
movies.info()
```

```python
dummies = movies["Genre"].str.get_dummies(", ")
dummies
```

```python
dummies.iloc[:10,:6]
```

```python
movies_windic = movies.join(dummies.add_prefix("Genero_"))
movies_windic
```

```python
movies_windic.iloc[782]
```

## Variables Continuas

```python
np.random.seed(12345)
```

```python
values = np.random.uniform(size=10)
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
```

```python
x = pd.get_dummies(pd.cut(values,bins))
x
```

```python
x = x.astype("uint8")
x
```

```python
z = pd.get_dummies(values)
z
```
