# Detección de Atípicos

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data.columns
```

```python
data.head()
```

```python
out = []
def Zscore_outlier(df,umbral):
  mean = np.mean(df)
  standarDesviation = np.std(df)
  for i in df:
      z = (i - mean) / standarDesviation
      if np.abs(z) > umbral:
        out.append(i)
  print("Outliers: ",out)
Zscore_outlier(data['LotArea'],umbral=3)
```

```python
Zscore_outlier(data['SalePrice'])
```

```python
data.dtypes
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def box_plots(df):
  plt.figure(figsize=(10,4))
  plt.title("Box Plot")
  sns.boxplot(df)
  plt.show()
```

## Ejemplo - Fireplaces

```python
box_plots(data['Fireplaces'])
```

```python
def hists_plot(df):
  plt.figure(figsize=(10,4))
  plt.hist(df)
  plt.title("Histograma Plot")
  plt.show()
```

```python
hists_plot(data['Fireplaces'])
```

```python
def dist_plot(df):
  plt.figure(figsize=(10,4))
  sns.distplot(df)
  plt.title("Distribution Plot")
  sns.despine()
  plt.show()
```

```python
dist_plot(data['Fireplaces'])
```

```python
from statsmodels.graphics.gofplots import qqplot
def qq_plots(df):
  plt.figure(figsize=(10,4))
  qqplot(df,line='s')
  plt.title("Normal QQPlot")
  plt.show()
```

```python
qq_plots(data['Fireplaces'])
```
