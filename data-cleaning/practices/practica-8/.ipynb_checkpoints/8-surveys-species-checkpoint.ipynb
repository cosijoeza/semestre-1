```python
import pandas as pd
```

## Lectura de bases de datos

```python
surveys_df = pd.read_csv("../../data/surveys.csv")
species_df = pd.read_csv("../../data/species.csv")
```

## Apila el segundo DataFrame debajo del primero

```python
survey_sub = surveys_df.head(10)
survey_sub
```

```python
survey_sub_last10 = surveys_df.tail(10)
survey_sub_last10
```

Restablecer los valores de índice al segundo dataframe

```python
survey_sub_last10 = survey_sub_last10.reset_index(drop=True)
survey_sub_last10
```

```python
new_survey = pd.concat([survey_sub,survey_sub_last10])
new_survey
```

```python
species_df.head()
```

## Inner join de las dos bases mediante la variable 'species_id'

```python
pd.merge(surveys_df, species_df, how="inner", on="species_id")
```

## Left join de las dos bases

```python
pd.merge(surveys_df, species_df, how="left")
```

## Right join de las dos bases

```python
pd.merge(surveys_df, species_df, how="right")
```

## Outer join de las dos bases

```python
pd.merge(surveys_df, species_df, how="outer")
```

## Group by 'species_id' (Surveys)

Obtenemos una base de datos que agrupa todas las claves de species_id y nos muestra la cantidad total que existe en cada una de las caracteristicas.

```python
df = surveys_df.groupby("species_id").count()
df.head(20)
```

## Exportar a CSV la mejor tabla

```python
dfToExport = pd.merge(surveys_df, species_df, how="left")
dfToExport.to_csv("../../data/surveysAndSpecies.csv", index=False)
```
