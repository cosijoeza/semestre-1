## Gráfica Frecuencia por Especie

```python
species_counts = surveys_df['species_id'].value_counts()
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10,4))
species_counts.plot(kind='bar')
plt.xticks(fontsize=6)
plt.title("Species ID")
plt.show()
```

## Gráfica del Promedio de Peso de las Especies por Sitio (Plot_ID)

```python
# Calculate the mean weight of each species by plot_id
average_weight_by_species_plot = surveys_df.groupby(['plot_id', 'species_id'])['weight'].mean().unstack()

# Plot the data
plt.figure(figsize=(14, 8))
average_weight_by_species_plot.plot(kind='bar', figsize=(14, 8), stacked=True)
plt.title('Promedio de Peso')
plt.xlabel('Plot_ID')
plt.ylabel('Mean Weight')
plt.legend(title='Especie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

## Gráfica del total de Machos vs Hembras

```python
# Calculate the total counts of males and females
sex_counts = surveys_df['sex'].value_counts()

# Plot the data
plt.figure(figsize=(8, 6))
sex_counts.plot(kind='bar', color=['red', 'pink'], width=0.6)
plt.title('Male vs Female')
plt.xlabel('Sexo')
plt.ylabel('Total')
plt.tight_layout()

# Show the plot
plt.show()
```

## Gráfico del peso total por Sexo para cada Sitio(Plot_ID)

```python
# Calculate the total weight by site (plot_id) and sex
total_weight_by_plot_sex = surveys_df.groupby(['plot_id', 'sex'])['weight'].sum().unstack()

# Plot the stacked bar chart
plt.figure(figsize=(12, 8))
total_weight_by_plot_sex.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Peso Total por Sexo para cada Sitio')
plt.xlabel('Plot_ID')
plt.ylabel('Weight')
plt.legend(title='Sexo', loc='upper right')
plt.tight_layout()

# Show the plot
plt.show()
```

## ¿Qué ocurre al ejecutar los siguientes códigos?:

```python
surveys_df[0:1]
```

```python
surveys_df[:4]
```

```python
surveys_df[:-1]
```

```python
surveys_df.iloc[0:4, 1:4]
```

```python
surveys_df.iloc[0:4, [1,4]]
```

 surveys_df[0:1]: Nos muestra un solo renglón y todas las columnas del dataset. 

 surveys_df[:4]: Nos muestra los renglones del 0 al 3 y todas las columas del dataset.

 surveys_df[:-1]: Nos muestra los ultimos renglones del dataset.

 dat.iloc[0:4, 1:4]: Toma las filas del 0 al 3 y las columnas del 1 al 4-1.

 
 dat.loc[0:4, [1,4]]: Toma los renglones del 0 al 3 y solo la columa 1 y la 4.

La diferencia es que con loc pedimos exactamente las columnas que necesitamos en cambio en iloc podemos seleccionar un conjunto de columas e indicar desde donde comienza hasta donde termina.
