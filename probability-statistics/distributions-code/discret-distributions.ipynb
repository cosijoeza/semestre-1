```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Distribución Binomial
Lanzamos una moneda 10 veces, esto lo repetimos 10000 veces. La probabilidad de éxito es 0.5

```python
binomial_samples = np.random.binomial(10, 0.5, size=10000)
print(binomial_samples)
```

## Grafica

```python
fig, ax = plt.subplots()
plt.hist(binomial_samples, density=True, bins=11)
# Labels
plt.title("Distribución binomial", size=14)
plt.xlabel("Muestras")
plt.ylabel("Densidad")
# Draw lines
plt.axvline(10 * 0.5, color="#000", linestyle="--")
plt.axvline(0, color="#000", linestyle="--")
plt.axvline(10, color="#000", linestyle="--")
plt.show()
```

# Distribución Hipergométrica

*Nota:*

- **k** = n (Exitos)
- **x** = k (Exito a calcular)
- **N** = M (Conjunto Total)
- **n** = N (Subconjunto)

```python
N = 8
n = 3
k = 5
x = 3
```

## Probabilidad Puntual
P(X = 3)

```python
# Do not change this code !
x = 3
stats.hypergeom.pmf(k = x,M = N, n = k, N = n)
```

## Probabilidad Acumulada
P(X <= 2)

```python
x = 2
stats.hypergeom.cdf(k = x,M = N, n = k, N = n)
```

## Probabilidades de un conjunto o rango

```python
X = np.arange(0,k+1)
hipergeo = stats.hypergeom(M=N, n=k, N=n)
probabilities = hipergeo.pmf(X)
print(X)
print(probabilities)
print(sum(probabilities))
```

## Grafica

```python
fig,ax = plt.subplots()
plt.bar(X,probabilities)
plt.title("Distribución Hipergeométrica")
plt.ylabel("Probabilidad")
plt.xlabel("Valores")
plt.show()
```

# Distribución de Poisson

## Probabilidad Puntual
P(X = 18)

```python
lamda = 3
success = 18
p_x = stats.poisson.pmf(k=success,mu=lamda)
print(p_x)
```

## Probabilidad Acomulada
*P(X <= 4)*

```python
p_x = stats.poisson.cdf(k=4,mu=3)
print(p_x)
```

```python
print(sum(stats.poisson.pmf(k=i,mu=3) for i in range(5)))
```

P(X > 4)

```python
p_x = 1 - stats.poisson.cdf(k=4,mu=3)
print(p_x)
```

```python
print(1-sum(stats.poisson.pmf(k=i,mu=3) for i in range(5)))
```

## Gráfica

```python
n = 19
X_poisson = []
lamda = 3
```

```python
X = np.arange(n)
print(X)
```

```python
for i in range(n):
  X_poisson.append(stats.poisson.pmf(k=i,mu=lamda))
print(X_poisson)
```

```python
plt.bar(X,X_poisson)
plt.xlabel('Exitos')
plt.ylabel('Probabilidad')
plt.title('Distribución Poisson')
plt.show()
```
