```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramtros
num_samples = 10000
```

```python
# Distribucion normal
uniform_samples = np.random.uniform(low=2, high=4, size=num_samples)
```

```python
# Distribucion normal
normal_samples = np.random.normal(loc=2, scale=1.0, size=num_samples)
```

```python
# Distribucion exponencial
exponential_samples = np.random.exponential(scale=3, size=num_samples)
```

## Grafica de las distribuciones

```python
plt.subplot(1, 3, 1)
plt.hist(uniform_samples, bins=30, color="skyblue", alpha=0.7)
plt.title("Distribucion uniforme")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
```
